<h2>1. Introduction</h2>

<p>
This project presents an in silico bioinformatics workflow for identifying differentially expressed genes (DEGs), hub genes, enriched biological pathways, and potential therapeutic targets associated with Lung Adenocarcinoma (LUAD).
</p>

<h2>2. Bioinformatics Pipeline</h2>

<div class="pipeline">

GEO → Quality Control → GEO2R → DEG Filtering → STRING →
Cytoscape → cytoHubba → Enrichr → KM Plotter → DGIdb

</div>

<h2>3. Differential Gene Expression</h2>

<p>
Genes were filtered according to statistical significance and fold change thresholds.
</p>

<div class="code">

Significantly Up-regulated

logFC ≥ 1.5
adj.P.Val < 0.05


Significantly Down-regulated

logFC ≤ -1.5
adj.P.Val < 0.05

</div>
