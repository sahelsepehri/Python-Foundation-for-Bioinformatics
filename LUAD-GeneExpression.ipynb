### Dataset Overview

> 1. The gene expression dataset **`GSE43458`** was obtained from the **`NCBI Gene Expression Omnibus (GEO)`** database. 
> 2. This dataset contains microarray gene expression profiles from **`110 human lung tissue samples`**, including **`80 lung adenocarcinoma (LUAD)`** samples and **`30 normal lung tissues`**.
> 3. The expression data were generated using the **`Affymetrix Human Gene 1.0 ST Array (GPL6244)`** platform.
> 4. The dataset was selected because it provides a sufficiently large number of tumor and normal samples for differential gene expression analysis.


| Feature            | Description                                      |
| ------------------ | ------------------------------------------------ |
| Dataset            | GSE43458                                         |
| Database           | NCBI GEO                                         |
| Organism           | Homo sapiens                                  |
| Platform           | GPL6244 (Affymetrix Human Gene 1.0 ST Array)     |
| Total Samples      | 110                                              |
| LUAD Samples       | 80                                               |
| Normal Samples     | 30                                               |
| Analysis Tool      | GEO2R                                            |
| Filtering Criteria | Adjusted P-value < 0.05, logFC > 1, logFC < -1 |


---

### Sample Group Definition (Tumor vs Normal)

> 1. Before performing differential expression analysis, the samples were assigned to two experimental groups in **`GEO2R`**.
> 2. The 80 lung adenocarcinoma samples were assigned to the **`Tumor group`**, while the 30 normal lung tissue samples were assigned to the **`Normal group`**.
> 3. These group definitions were used by GEO2R to compare gene expression levels between tumor and normal tissues.

---

### Differential Expression Analysis (GEO2R)


> 1. Differential expression analysis was performed using the **`GEO2R`** web tool to calculate log fold change (logFC)and adjusted P-values for each gene.
> 2. The resulting table was exported to **`Microsoft Excel`**, where statistical filtering was applied using an **`adjusted P-value < 0.05`**. Genes with **`logFC > 1`** were classified as **`up-regulated`**, whereas genes with **`logFC < -1`** were classified as **`down-regulated`**.
> 3. The filtered gene lists were used for subsequent analyses.




---

### GEO2R Output
> After applying the filtering criteria (adjusted P value < 0.05, logFC > 1, and logFC < -1), 319 upregulated genes and 706 downregulated genes were identified for further analysis.


| Category      | Genes |
| ------------- | ----: |
| Upregulated   |  319     |
| Downregulated |   706 |
| Total         |   1025 |





Volcano Plot Interpretation

> - The volcano plot shows the differential gene expression results between LUAD and normal lung tissues.
> - Genes on the right side are upregulated, while genes on the left side are downregulated.
> - Points higher on the plot have lower adjusted P values and are therefore more statistically significant.




MA Plot Interpretation

> - The MA plot displays the relationship between average gene expression and log2 fold change for all analyzed genes.
> - Most genes are centered around a log2 fold change of zero, indicating similar expression levels between the two groups.
> - Genes with significant differential expression appear above or below the center line and are highlighted according to their regulation status.
> - The overall distribution suggests that differential expression occurs across a wide range of expression levels rather than being limited to highly or weakly expressed genes.



Expression Density Plot Interpretation

> - The expression density plot illustrates the distribution of normalized expression values for all samples included in the analysis.
> - Each curve represents the expression profile of one sample.
> - The density curves show similar overall distributions across both tumor and normal samples, indicating that the expression data are generally comparable after normalization. This provides confidence that the samples are suitable for downstream differential expression analysis.

### 2. Protein-Protein Interaction (PPI) Network Construction (STRING & Cytoscape)

> 1. After identifying the differentially expressed genes (DEGs), the next step was to investigate how their encoded proteins interact with each other. Since genes rarely function independently, examining protein-protein interactions helps identify functional relationships and biological networks involved in the disease.

> 2. The filtered DEGs were uploaded to the `STRING` database to retrieve known and predicted `protein-protein interactions (PPIs)`. The resulting interaction network was then imported into `Cytoscape`, where the network was visualized and prepared for further topological analysis.


###  Protein-Protein Interaction (PPI) Network


---

### 3. Hub Gene Identification (cytoHubba)

> Not all genes within a protein interaction network contribute equally. Some proteins interact with many other proteins and may play central roles in regulating important biological processes. These highly connected genes are commonly referred to as `hub genes`.

> 1. To identify hub genes, the PPI network was analyzed using the `cytoHubba` plugin in Cytoscape.
> 2. The `Maximal Clique Centrality (MCC)` algorithm was used to rank genes according to their importance within the network.
> 3. Based on the MCC scores, the top 10 hub genes were selected for subsequent functional enrichment, survival, protein expression, and drug-gene interaction analyses.


STRING identified BUB1, MELK and CDK1 as central nodes.

Because all three genes were also enriched in Cell Cycle pathways and associated with poor survival in LUAD, we selected them for downstream drug-gene interaction analysis.

The hub genes identified in the PPI network were consistently enriched in cell cycle pathways and were associated with poor overall survival, suggesting their biological importance in LUAD.

### Results

> 1. The MCC algorithm identified ten highly connected hub genes within the PPI network: `BUB1, MELK, CDK1, PBK, CCNA2, KIF11, DLGAP5, BUB1B, ASPM, and CCNB1`.
> 2. These genes showed the highest network connectivity and were selected as candidate hub genes for subsequent analyses.
> 3. The first six genes received identical MCC scores, indicating that they occupy similarly central positions within the interaction network.
